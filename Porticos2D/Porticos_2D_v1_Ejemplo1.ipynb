# Geometria de los Elementos

```python
### Elementos Individuales

def crear_elemento(Id, E, A, I, xi, yi, xf, yf,             
                   giro_i, giro_f, desy_i, desy_f, desx_i, desx_f,
                   TC, W):
    
    ID = Id
    l = ((xf - xi) ** 2 + (yf - yi) ** 2) ** 0.5
    Cx = (xf - xi) / l
    Cy = (yf - yi) / l
    GDL = [giro_i, giro_f, desx_i, desx_f, desy_i, desy_f]
    
    if TC == 'Nada':
        
        Me_i = 0
        Me_f = 0
        Ve_i = 0 
        Ve_f = 0
    
    elif TC == 'Puntual1':
        
        Me_i = (W * l) / 8.0
        Me_f = - (W * l) / 8.0
        Ve_i = (W) / 2.0
        Ve_f = (W) / 2.0
    
    elif TC == 'Distr':

        Me_i = (W * l ** 2) / 12.0
        Me_f = - (W * l ** 2) / 12.0
        Ve_i = (W * l) / 2.0
        Ve_f = (W * l) / 2.0

    return ID, l, Cx, Cy, A, E, I, GDL, Me_i, Me_f, Ve_i, Ve_f 
```

# Matriz de Rigidez

### Matriz de la Rigidez en Coordenadas Locales

```python
def matriz_local(L, A, E, I):
    
    K11 = (4 * E * I) / L 
    K21 = (2 * E * I) / L
    K31 = (K11 + K21) / L
    K41 = -(K11 + K21) / L
    K51 = 0.0
    K61 = 0.0
    
    K12 = K21
    K22 = (4 * E * I) / L
    K32 = (K22 + K21) / L
    K42 = -(K22 + K21) / L
    K52 = 0.0
    K62 = 0.0
    
    K13 = K31
    K23 = K32
    K33 = (K11 + K22 + 2 * K21) / (L ** 2) 
    K43 = -(K11 + K22 + 2 * K21) / (L ** 2) 
    K53 = 0.0
    K63 = 0.0
    
    K14 = K41
    K24 = K42
    K34 = K43 
    K44 = (K11 + K22 + 2 * K21) / (L ** 2)
    K54 = 0.0
    K64 = 0.0
    
    K15 = K51
    K25 = K52
    K35 = K53 
    K45 = K54
    K55 = A * E / L 
    K65 = - A * E / L
    
    K16 = K61
    K26 = K62
    K36 = K63 
    K46 = K64
    K56 = K65 
    K66 = A * E / L
    
    k = [[K11, K21, K31, K41, K51, K61],[K12, K22, K32, K42, K52, K62],[K13, K23, K33, K43, K53, K63],
        [K14, K24, K34, K44, K54, K64],[K15, K25, K35, K45, K55, K65],[K16, K26, K36, K46, K56, K66]]
    
    matriz = np.matrix(np.round(k,2))
        
    return matriz
```

```python
Matriz_local = [None] * len(Tabla_Datos)

for i in np.arange(len(Tabla_Datos)):

    L = Tabla_Datos['L'][i]
    A = Tabla_Datos['A'][i]
    E = Tabla_Datos['E'][i]
    I = Tabla_Datos['I'][i]

    Matriz_local[i] = matriz_local(L, A, E, I)
    
for i in np.arange(len(Matriz_local)):
    
    print 'Matriz del Elemento ' + Tabla_Datos['ID'][i]
    print Matriz_local[i]
    print '\n'
```

### Matriz de Transformación

```python
def matriz_T(CX, CY):
    
    T11 = 1.0 
    T21 = 0.0
    T31 = 0.0
    T41 = 0.0
    T51 = 0.0
    T61 = 0.0
    
    T12 = 0.0
    T22 = 1.0
    T32 = 0.0
    T42 = 0.0
    T52 = 0.0
    T62 = 0.0
    
    T13 = 0.0
    T23 = 0.0
    T33 = -1.0 * CY 
    T43 = 0.0 
    T53 = CX
    T63 = 0.0
    
    T14 = 0.0
    T24 = 0.0
    T34 = 0.0
    T44 = -1.0 * CY
    T54 = 0.0
    T64 = CX
    
    T15 = 0.0
    T25 = 0.0
    T35 = CX 
    T45 = 0.0
    T55 = CY 
    T65 = 0.0
    
    T16 = 0.0
    T26 = 0.0
    T36 = 0.0
    T46 = CX
    T56 = 0.0 
    T66 = CY
    
    T = [[T11, T21, T31, T41, T51, T61], [T12, T22, T32, T42, T52, T62], [T13, T23, T33, T43, T53, T63],
         [T14, T24, T34, T44, T54, T64], [T15, T25, T35, T45, T55, T65], [T16, T26, T36, T46, T56, T66]]
    
    matriz = np.matrix(np.round(T,4))
        
    return matriz
```

```python
Matriz_T = [None] * len(Tabla_Datos)

for i in np.arange(len(Tabla_Datos)):

    CX = Tabla_Datos['CX'][i]
    CY = Tabla_Datos['CY'][i]
    
    Matriz_T[i] = matriz_T(CX,CY)
    
for i in np.arange(len(Matriz_T)):
    
    print u'Matriz de Transformación ' + Tabla_Datos['ID'][i]
    print Matriz_T[i]
    print '\n'
```

### Matriz de la Rigidez en Coordenadas Globales

```python
Matriz_global = [None] * len(Tabla_Datos)

for i in np.arange(len(Tabla_Datos)):

    Matriz_global[i] = Matriz_T[i] * Matriz_local[i] * Matriz_T[i]
    
    print u'Matriz del Elemento ' + Tabla_Datos['ID'][i] + ' (Globales)'
    print Matriz_global[i]
    print '\n'
```

### Matriz de Rigidez de la Estructura (GDL)

```python
N_GDL = 6
Matriz_GDL = [[None] * N_GDL for i in range(N_GDL)]

N = np.arange(1, N_GDL + 1)

for i in np.arange(len(N)):
    for j in np.arange(len(N)):
        Matriz_GDL[i][j] = str(N[i]) + str(N[j])

Matriz_GDL = np.matrix(Matriz_GDL)
Matriz_GDL
```

```python
gdl_sel = []
rig_sel = []

for e in np.arange(len(Tabla_Datos)):
    
    for g in Matriz_GDL.flat:

        cuenta = -1
        for i in Tabla_Datos['GDL'][e]:

            for j in Tabla_Datos['GDL'][e]:

                cuenta = cuenta + 1

                if str(i) + str(j) == g:

#                     print str(i) + str(j), cuenta, Matriz_global[e].flat[cuenta]
                    
                    gdl_sel.append(Matriz_global[e].flat[cuenta])
                    rig_sel.append(str(i) + str(j))
            
Matriz_Ppal = np.matrix([np.zeros(N_GDL) for i in range(N_GDL)])
Matriz_Ppal

for g in np.arange(len(np.matrix(Matriz_GDL).A1)):
        
    for i in np.arange(len(rig_sel)):
        
        if Matriz_GDL.flat[g] == rig_sel[i]:
                    
            Matriz_Ppal.flat[g] = Matriz_Ppal.flat[g] + gdl_sel[i]  
            
# Matriz_K = Matrix(Matriz_Ppal)
# Matriz_Kprint 'Resultados de los Mu_global'
```

```python
Columns_GDL = ['$\Theta_1$', '$\Theta_2$', '$\Theta_3$', '$\Theta_4$',
               '$\Delta_1$', '$\Delta_2$']

Matriz_GDL = pd.DataFrame(data = Matriz_Ppal,index = Columns_GDL, columns = Columns_GDL)
Matriz_GDL
```

### Grados de Libertad de Cada Elemento

```python
GLs = [None] * len(Tabla_Datos)
for i in np.arange(len(Tabla_Datos)):
    
    Datos = Tabla_Datos['GDL'][i]
    GL = np.matrix([[0.0] * len(Datos)]).T
    for j in np.arange(len(Datos)):
        
        if Datos[j]-1 == -1:
            GL[j] = 0
        
        else:
            GL[j] = R[Datos[j]-1]

    GLs[i] = GL

Me_global = [None] * len(Tabla_Datos)
for i in np.arange(len(Tabla_Datos)):
    
    Datos = Tabla_Datos['GDL'][i]
    Me_gl = np.matrix([[0.0] * len(Datos)]).T
    
    Me_gl[0] = Tabla_Datos['Me_i'][i]
    Me_gl[1] = Tabla_Datos['Me_f'][i]
    Me_gl[4] = Tabla_Datos['Ve_i'][i]
    Me_gl[5] = Tabla_Datos['Ve_f'][i]
    
    Me_global[i] = Me_gl
    
Mu_global = [None] * len(Tabla_Datos)
for i in np.arange(len(Tabla_Datos)):
    
    Mu_global[i] = Matriz_global[i] * GLs[i]
```

### Momentos maximos en el intermedio de los elementos

```python
def moment_max_dist(M_i, M_f, V_i, V_f, L, W):
    
    x = (V_i * L) / (V_i + V_f)
    
    M_max = - M_i + V_i * x - (W * (x ** 2) / 2.0)
    
    return np.round(x,4), np.round(M_max,4)
```
